Exercise4: Write a Python program to find the median of a given array of numbers with for and if loop

```python
xsort = np.sort(x)
mid = int(len(x)/2)
if xsort.size%2==0: 
    median = (xsort[mid]+xsort[mid+1])/2
else: median = xsort[mid]
print("Median: " + str(median))
```
